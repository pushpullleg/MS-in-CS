Pivot and pivot table

Basically  pivot is a method on the DataFrame object,there are arguments such as index,column,values


```python
import numpy as np
import pandas as pd
df=pd.read_csv('weather_by_cities.csv', parse_dates=['day'])
df
```

```python
p=df.pivot(index='temperature',columns='city')
p
```

## Difference between arguments and parameters

Parameter

Definition: A parameter is a variable in the declaration of a function or method. It defines what kind of arguments the function or method can accept.

Example: In the function definition def add(a, b):, a and b are parameters.

Argument
Definition: An argument is the actual value that is passed to the function or method when it is called. Arguments are the real data you pass to the function's parameters.


Example: In the function call add(2, 3), 2 and 3 are arguments.

## What is the difference between pivot and pivot table?

Basically aggreagation is the major difference, and also margins.
other difference are handling missing values. 
parameters: fill_value,margins, aggfunction

```python
pt = df.pivot_table(index='city', columns='day', values='temperature', aggfunc=np.mean)
pt
```

## What if there is many dates, you would like to find the avg temperature for the month?


The Best Way to do it converting the day/date column into time_stamp , you can use function or parse_dates while reading CSV.

And Use the Grouper Function . It has got parameter called frequency , and Key works on the column 

note it is not the method or argument.hence need to to use pd.Grouper(key=,freq=)

```python
k = df.pivot_table(index=pd.Grouper(key='day', freq='M'), values='temperature', aggfunc=np.mean)
k
```
