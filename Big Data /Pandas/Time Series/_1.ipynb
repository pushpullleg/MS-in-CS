```python
## If we want to know about the monthly data , not intersted in the daily data

## We use resasmple method 

df.Open.resample('ME').mean()
```

```python
df.resample('QE').max()
```

```python
## Here are some common frequency strings you can use with the resample method:

'D': Calendar day frequency
'W': Weekly frequency
'M': Month end frequency
'Q': Quarter end frequency
'A': Year end frequency
'H': Hourly frequency
'T' or 'min': Minute frequency
'S': Second frequency
```

Note

The resample method in pandas requires the DataFrame to have a DateTime index.
