## how to simply replace some cells value to desired one ?

```python
import numpy as np
df.replace(-99999,np.NaN)
```

```python
df.replace({
    'temperature':-99999,
    'event':'0'
},np.NaN)
```

## If there are units in the cell value, and you have to remove it , how do go about it?


By using regex 

```python
df1=df.replace('[A-Za-z]','',regex=True)
df1
```

Without regex=True: The pattern [A-Za-z] is treated as a literal string, and no replacements are made.

df1 = df.replace('[A-Za-z]', '')

With regex=True: The pattern [A-Za-z] is interpreted as a regular expression, and all alphabetic characters are replaced with an empty string.

df1 = df.replace('[A-Za-z]', '', regex=True)


## When you wanted to to remove only replace specific columns 

```python
df1=df.replace({
    'temperature': '[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'', regex=True)
df1
```
