## Differences Between pivot, melt, and stack in Pandas


pivot: Reshapes data from long to wide format.

Example: df.pivot(index='date', columns='city', values='temperature')

melt: Reshapes data from wide to long format.

Example: pd.melt(df, id_vars='day', var_name='city', value_name='temperature')

stack: Stacks columns into a MultiIndex Series.

Example: df.stack()
